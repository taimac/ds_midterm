# Sales Analysis of Steel Products

The idea of this project is to analyze the sales behavior of my commercial representation office. Therefore, the data is mine and was extracted from my invoice system and exported to a CSV file. I operate in the field of sales of carbon flat steels, distributing materials to an important consumption region in southern Brazil. The names of the clients have been excluded to ensure the privacy of data from other companies. With this, what I intend to do here is an initial exploration of the sales performance of my office based on the analysis of the quantity of materials sold, product families, sales behavior over time, price variations, and more.

```python
# Start Web Scraping
r = requests.get("http://www.ipeadata.gov.br/ExibeSerie.aspx?serid=38590&module=M")
```

```python
# Convert to a beautiful soup object
soup = bs(r.content)
```

```python
today  = pd.Timestamp.today()
final_date = '01/01/2008'
final_date = datetime.strptime(final_date, '%d/%m/%Y')
range_date = (today - final_date).days + 1
range_date
```

```python
range_date = today - final_date
range_date = range_date.days
range_date = range_date + 1
range_date
```

```python
def get_dollar_value():
    date = []
    value = []
    for id in range(range_date):
        id = str(id)
        pattern = 'grd_DXDataRow' + id
        #tr_list.append(pattern)
        id = int(id)
        tr =  soup.find('tr', {'id' : pattern})
        t_data_tags = tr.find_all("td")
        date.append(t_data_tags[0].text)
        value.append(t_data_tags[1].text)
        date[id] = datetime.strptime(date[id], '%d/%m/%Y')
        if (date[id]) == final_date:
            break
        
    data = {'Date': date, 'US_Dollar_in_R$': value}
    dollar_df = pd.DataFrame(data)
 
    return dollar_df
```

```python
dollar_df = get_dollar_value()
```
