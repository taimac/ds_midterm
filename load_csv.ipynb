```python
# Import information about sales invoices
import pandas as pd
csv_files = ['./data/csv_files/2014.csv','./data/csv_files/2015.csv',
         './data/csv_files/2016.csv','./data/csv_files/2017.csv', './data/csv_files/2018.csv',
         './data/csv_files/2019.csv','./data/csv_files/2020.csv','./data/csv_files/2021.csv',
        './data/csv_files/2022.csv','./data/csv_files/2023.csv','./data/csv_files/2024.csv']

dataframes = [pd.read_csv(file_path) for file_path in csv_files]

dataframes

sales_data = pd.concat(dataframes, ignore_index=True)

sales_data 

sales_data.to_csv('data/csv_files/sales_data.csv', index = False)

sales_data
```
